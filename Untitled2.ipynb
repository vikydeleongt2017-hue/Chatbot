```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd

tree = {
    "start": {
        "message": "🤖 Dupie: ¡Hola! ¿En qué puedo ayudarte?",
        "options": {
            "📚 eTrámites": "etramites",
            "💳 Pagos": "pagos",
            "🎓 Portal Académico": "portal"
        }
    },
    "etramites": {
        "message": "¿Qué trámite necesitas?",
        "options": {
            "📄 Certificaciones": "cert",
            "⬅ Volver": "start"
        }
    },
    "cert": {
        "message": "Puedes solicitar certificaciones en el portal.",
        "options": {
            "⬅ Volver": "etramites"
        }
    }
}

current_node = "start"
chat_box = widgets.VBox()
log = []

def add_message(text, sender="bot"):
    style = {
        "bot": "background:#eee;padding:8px;border-radius:10px;margin:5px;",
        "user": "background:#1D9E75;color:white;padding:8px;border-radius:10px;margin:5px;text-align:right;"
    }
    chat_box.children += (widgets.HTML(f"<div style='{style[sender]}'>{text}</div>"),)

def render(node_id):
    global current_node
    current_node = node_id
    node = tree[node_id]
    
    add_message(node["message"], "bot")
    
    buttons = []
    for label, next_node in node["options"].items():
        btn = widgets.Button(description=label)
        
        def on_click(b, label=label, next_node=next_node):
            add_message(label, "user")
            log.append({"node": current_node, "choice": label})
            render(next_node)
        
        btn.on_click(on_click)
        buttons.append(btn)
    
    display(widgets.VBox(buttons))

display(chat_box)
render("start")
```
